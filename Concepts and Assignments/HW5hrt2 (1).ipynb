```python
with open('/content/drive/MyDrive/data5.pkl', 'rb') as file:
  data_loaded = pickle.load(file)

X = data_loaded['X']
y = data_loaded['y']
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(X)

# Calculate sum of squared distances of samples to their closest cluster center
clustering_error = kmeans.inertia_
print("Clustering Error (Sum of Squared Distances):", clustering_error)

# Obtain cluster labels from K-means
cluster_labels = kmeans.labels_

# Calculate adjusted Rand score
adjusted_rand_score_value = adjusted_rand_score(y, cluster_labels)
print("Adjusted Rand Score:", adjusted_rand_score_value)
```

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative_clustering = AgglomerativeClustering(n_clusters=10)
cluster_labels = agglomerative_clustering.fit_predict(X)

# Calculate adjusted Rand score
adjusted_rand_score_value = adjusted_rand_score(y, cluster_labels)
print("Adjusted Rand Score (Hierarchical Clustering):", adjusted_rand_score_value)
```
