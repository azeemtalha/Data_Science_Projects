Question 3

```python
dataset = [55, 60, 65, 70, 75, 80, 85, 90, 95, 200]
mean = np.mean(dataset)
std_dev = np.std(dataset)

z_scores = [(x - mean) / std_dev for x in dataset]

outliers = [x for x, z in zip(dataset, z_scores) if z > 2 or z < -2]


print("Z-scores:")
for x, z in zip(dataset, z_scores):
    print(f"Data point {x}: Z-score = {z}")

print("\nOutliers:")
print(outliers)
```

```python
Q1 = np.percentile(dataset, 25)
Q3 = np.percentile(dataset, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

IQR_outliers = [x for x in dataset if x < lower_bound or x > upper_bound]

print("IQR Outliers:")
print(IQR_outliers)
```

```python
median = np.median(dataset)

abs_deviations = np.abs(dataset - median)

MAD = np.median(abs_deviations)

outlier_boundary = 2 * MAD

MAD_outliers = [x for x in dataset if np.abs(x - median) > outlier_boundary]

print("MAD Outliers:")
print(MAD_outliers)
```

Question 4

```python
from scipy.stats import ttest_ind

# Group A data
group_a = [50, 60, 40, 45, 35]

# Group B data
group_b = [35, 40, 30, 45, 50]

# Perform independent two-sample t-test
t_statistic, p_value = ttest_ind(group_a, group_b)

# Set significance level
alpha = 0.05

# Print results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference between the means of the two groups.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference between the means of the two groups.")
```

Question 5

```python
from scipy.stats import f_oneway
group1 = [12, 15, 14, 11, 13]
group2 = [22, 20, 21, 19, 23]
group3 = [32, 30, 28, 33, 31]

f_statistic, p_value = f_oneway(group1, group2, group3)

alpha = 0.05

print("One-way ANOVA Test Results:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There are statistically significant differences among the means of the three groups.")
else:
    print("Fail to reject the null hypothesis. There are no statistically significant differences among the means of the three groups.")
```

```python
tips = sns.load_dataset("tips")

# Create a FacetGrid and KDE plots
g = sns.FacetGrid(data=tips, col="day", hue="smoker", aspect=2)
g.map(sns.kdeplot, "total_bill", fill=True, alpha=0.6)

# Customize the plot
g.fig.suptitle("Total Bill Distribution by Day and Smoker Status", y=1.05)
g.set_axis_labels("Total Bill", "Density")

# Rotate x-axis labels for better readability
g.fig.subplots_adjust(bottom=0.2)

# Show the plot
plt.show()
```

```python
tips = sns.load_dataset("tips")

# Create a FacetGrid with KDE plots for total bill distribution, differentiated by smoker status and day
g = sns.FacetGrid(tips, col="day", hue="smoker", col_wrap=2, height=4, aspect=1.5)
g.map(sns.kdeplot, "total_bill", shade=True)

# Set aspect ratio for each plot
g.set(ylim=(0, 0.1))

# Add a main title for the FacetGrid
g.fig.suptitle('Total Bill Distribution by Day and Smoker Status', fontsize=16)

# Add legend
g.add_legend()

# Show the plot
plt.show()
```
